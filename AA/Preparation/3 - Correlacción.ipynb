### Importa los módulos necesarios y establece configuraciones

```python
import pandas as pd
import matplotlib.pyplot as plt
import pandas as pd
import funs
import seaborn as sns
import plotly.plotly as py
from plotly.graph_objs import *
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
init_notebook_mode(connected=True)
```

### Importa los nodos y sus respectivos periodos

```python
tabla={}
for T in [1,2,3]:
    tabla[T]=[]
    for nodo in [2,3,5]:
        data = pd.read_csv('nodo%d_periodo%d.csv'%(nodo,T),parse_dates=True,index_col=0)
        data.columns = [j+str(nodo) for j in data.columns]
        tabla[T].append(data)
```

## Concatena los DFs contenidos en tabla

```python
periodo = {}
for T in [1,2,3]:
    periodo[T] = tabla[T][0].join(tabla[T][1], how='outer').join(tabla[T][2], how='outer')
```

## Ordena las columnas por orden alfabético

```python
for T in periodo:
    periodo[T] = periodo[T].reindex(columns = sorted(periodo[T].columns))
```

## Matrices de correlación entre variables de diferentes nodos

```python
for T in periodo:
    tcorr = periodo[T].corr()
    plt.figure(figsize=[11,8])
    sns.heatmap(tcorr.iloc[3::,3::], annot=True, cmap='RdYlGn',fmt=".2f")
    plt.show()
```

La correlación entre las variables se hace más debil entre mayor es la cantidad de días en los que se registran valores.
