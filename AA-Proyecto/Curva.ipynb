```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Tus datos
mediciones_cacao = pd.read_csv("medidas.csv")
mediciones_cacao = mediciones_cacao.query('NODO == 2').reset_index(drop=True)

x = range(len(mediciones_cacao['INTENSIDAD LUMÍNICA'][500:1600]))
y = mediciones_cacao['INTENSIDAD LUMÍNICA'][500:1600].copy()
sols = {}
grado=15
z = np.polyfit(x, y, grado, full=True)
sols[grado] = z

# Pintar datos
plt.plot(x, y, 'o')

# Pintar curvas de ajuste


coefs= z[0]
p = np.poly1d(coefs)
plt.plot(range(len(y)), p(range(len(y))), "-", label="Gr: %s. " % (grado) )
plt.legend()
```
