```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc

fig, ax1 = plt.subplots()
ax2 = ax1.twinx()
ax3 = ax1.twinx()
x1,=ax1.plot([1,3],[4,2],color='r',label='1')
x2,=ax2.plot([3,8],[6,3],color='y',label='12')
x3,=ax3.plot([4,8],[1,3],color='k',label='13')


ax1.set_ylabel("Density")
ax2.set_ylabel("Temperature")
ax3.set_ylabel("Velocity")

lns = [x1, x2, x3]
ax1.legend(handles=lns)
# right, left, top, bottom
ax3.spines['right'].set_position(('outward', 60))      
# no x-ticks                 
ax3.xaxis.set_ticks([])
# Sometimes handy, same for xaxis
#par2.yaxis.set_ticks_position('right')
ax1.yaxis.label.set_color(x1.get_color())
ax2.yaxis.label.set_color(x2.get_color())
ax3.yaxis.label.set_color(x3.get_color())
```
