```python
wind= 100 # Acá se define el tamaño de la ventana, cada cuanto se mueve :D!
sigma=1.5 #que dice que fuera de dos ya es valor anormal
# Límite superior de la ventana desliznte definido por Tukey
Nodo2['up']= Nodo2.TEMPERATURA.rolling(window=wind).quantile(.25)-(sigma*Nodo2.TEMPERATURA.rolling(window=wind).std())
# Límite inferior de la ventana desliznte definido por Tukey
Nodo2['down']= Nodo2.TEMPERATURA.rolling(window=wind).quantile(.75)+(sigma*Nodo2.TEMPERATURA.rolling(window=wind).std())
# Grafica de Temperatura, up, Down, respecto a TIMe
Nodo2.iloc[:,[2,7,8,9]].plot(x='TIME')
```
